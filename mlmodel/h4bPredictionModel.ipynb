```python
import numpy as np
import pandas as pd
```

```python
trainSet = pd.read_csv('Training.csv')
testSet = pd.read_csv('Testing.csv')
```

```python
trainSet.head()
```

```python
trainSet.describe
```

```python
trainSet = trainSet.loc[:, ~trainSet.columns.str.contains('^Unnamed')]
trainSet.head()
```

```python
trainSet.describe
```

```python
trainSet.shape
```

```python
trainSet.isnull().sum()
```

```python
testSet = testSet.loc[:, ~testSet.columns.str.contains('^Unnamed')]
testSet.head()
```

```python
testSet.shape
```

```python
testSet.describe
```

```python
x_train = trainSet[trainSet.columns.difference(['prognosis'])]
y_train  = trainSet[['prognosis']]
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
x_test= testSet[testSet.columns.difference(['prognosis'])]
y_test  = testSet[['prognosis']]
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
clf =  DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score = accuracy_score(y_test, y_pred)
print(score)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf= RandomForestClassifier(random_state=42)
model_rf = rf.fit(x_train,y_train)
tr_pred_rf = model_rf.predict(x_train)
ts_pred_rf = model_rf.predict(x_test)

print("training accuracy is:",accuracy_score(y_train,tr_pred_rf))
print("testing accuracy is:",accuracy_score(y_test,ts_pred_rf))
```

```python
import pickle
with open('classifier.pkl','wb') as file:
  pickle.dump(clf, file)
```
